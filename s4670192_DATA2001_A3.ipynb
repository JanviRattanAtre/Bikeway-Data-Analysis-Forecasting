## 2016

```python
# Initialize the 'Total Cycle Traffic' column with zeros
df_2016['Total Cycle Traffic'] = 0.0

# Iterate through the columns in df_2016
for column in df_2016.columns:
    # Check if the column name contains 'Cyclists Inbound'
    if 'Cyclists Inbound' in column:
        # Find the index of the current column
        column_index = df_2016.columns.get_loc(column)
        
        # Calculate the 'Total Cycle Traffic' by adding the (Inbound) value in the current column
        # with the (Outbound) value in the immediately following column in the same row
        inbound_column = df_2016.iloc[:, column_index]
        outbound_column = df_2016.iloc[:, column_index + 1]
        
        # Replace NaN values with 0
        inbound_column = inbound_column.fillna(0)
        outbound_column = outbound_column.fillna(0)
        
        df_2016['Total Cycle Traffic'] += inbound_column + outbound_column

```

```python
# Extract the "Total Cycle Traffic"" column from df_2016
total_cycle_traffic_column = df_2016['Total Cycle Traffic']

# Concatenate the "Total Cycle Traffic"" column with columns_2016_extracted
columns_2016_extracted = pd.concat([columns_2016_extracted, total_cycle_traffic_column], axis=1)
```

## b.1 Isolate the trend

```python
# We have 12 months in a year, so roll to 1 for one month and 12*5=60 for 5 years.
combined_df['MA-12'] = combined_df['Total Cycle Traffic'].rolling(12).mean()
combined_df['MA-60'] = combined_df['Total Cycle Traffic'].rolling(60).mean()
combined_df[['Total Cycle Traffic','MA-12','MA-60']].plot(figsize=(24, 6))
```

**OBSERVATION**<BR>
* MA-12 represents a 12-month (1-year) rolling moving average. It calculates the average Total Cycle Traffic over the past 12 months at each data point. This moving average helps smooth out short-term fluctuations and highlights longer-term trends or patterns.

* MA-60 represents a 60-month (5-year) rolling moving average. It calculates the average Total Cycle Traffic over the past 60 months at each data point. This moving average provides an even smoother representation of the data, further reducing noise and emphasizing long-term trends.

The line plot displays three lines:

* 'Total Cycle Traffic': This shows the actual data for Total Cycle Traffic over time.<br><br>
* 'MA-12': This line represents the 12-month rolling moving average, helping visualize the short to medium-term trends.<br><br>
* 'MA-60': This line represents the 60-month rolling moving average, emphasizing long-term trends.
The purpose of using moving averages is to make underlying trends more apparent by removing short-term noise or fluctuations. In the graph, you can observe how 'MA-12' and 'MA-60' provide a smoother view of the traffic data, making it easier to identify trends and patterns over different time scales.

* Root Mean Square Error (RMSE): RMSE is a slightly more complex metric. It first calculates the squared differences between actual and predicted values, then takes the square root of the average of those squared differences. In our case, the RMSE is approximately 15,016.19. Like MAE, RMSE provides a measure of the model's prediction error, but it gives more weight to larger errors. It's in the same units as the predicted values.

In summary, your forecasting model has an MAE of 13,056.76, which means that, on average, the absolute difference between your predictions and the actual values is about 13,056.76 units. The RMSE of 15,016.19 accounts for both the magnitude and the square of errors, and it can be used to assess the goodness of fit of your model. A lower RMSE indicates a better fit, similar to MAE.
